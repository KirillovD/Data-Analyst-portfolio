## GDP DS
### This dataset contains GDP data for every country from 1999 to 2022 in billions of US dollars. It is taken from https://www.kaggle.com/code/alejopaullier/gdp-by-country-1999-2022/data?scriptVersionId=105241333. This dataset has missing values for some countries, as well as for the whole year 2011. It is not the best choice in sence of informational content of the dataset. But the more problem we have got,  the better this dataset is to practice data cleaning.

```python
gdp = pd.read_csv(r"C:\Users\1\Desktop\GDP by Country 1999-2022.csv", index_col=[0])
gdp
```

#### Let's drop the missing and zero values

```python
gdp=gdp.dropna()
gdp = gdp.loc[(gdp != "0").all(axis=1)]
gdp.shape
```

#### The large numbers are save in DS with comma and dot like this 1, 123. 789. This format can not be converted into float. Let's delete the commas to solve this problem. At the end we will drop the columns after the year 2018, because we don't have this imformation in population DS

```python
for col in gdp.columns:
    gdp[col]=gdp[col].astype(str).str.replace(",","")
    gdp[col]=gdp[col].astype(float)
gdp = gdp.iloc[:,:-4]
```

#### Let's define the fuction that will melt a dataframe, to produce one column for GDP and Population from two datasets

```python
def melt_and_sort(df,col_name):
    if df.index.name == None:
        df.index.name = "Country"
    df = df.reset_index(level=0)
    df = pd.melt(df, id_vars = "Country", value_vars = df.columns, 
                          var_name = "Year", value_name = col_name)
    df = df.sort_values(["Country","Year"])
    df= df.set_index(["Country"])
    return df
gdp_merge = melt_and_sort(gdp_merge, "GDP_bln_USD")
population_merge= melt_and_sort(population_merge, "Population")
```

```python
gdp = pd.concat([gdp_merge, population_merge["Population"]], axis=1)
```

```python
gdp = gdp.reset_index(level=0)
gdp["GDP_bln_USD"]=round(gdp["GDP_bln_USD"],2)
```

#### Now we have the DS with only four columns. Let's calculate on it the growth rate of GDP, GDP per capita and additionaly growth rate of GDP per capita. We can do it by adding the new columns to gdp dataset. For growth rate we will use the pct_change() method

```python
gdp
```

```python
gdp["GDP_per_Capita"]=(gdp["GDP_bln_USD"]*(10**9)) / (gdp["Population"]*(10**3))
gdp["GDP_per_Capita"]=gdp["GDP_per_Capita"].astype(int)
```

```python
gdp.head()
```

```python
gdp["Growth_rate"] = round(gdp["GDP_bln_USD"].pct_change()*100,2)
```

```python
gdp.head()
```

```python
gdp["growth_per_capita"]=round(gdp["GDP_per_Capita"].pct_change()*100,2)
```
